```python
import pandas as pd
from matplotlib import pyplot as plt

sales_data = {
    "Quarter":["Q1 2022","Q2 2022","Q3 2022","Q4 2022","Q1 2023"],
     "Item A":[12,20,50,80,10],
     "Item B":[12,14,20,49,80],
     "Item C":[26,69,52,36,59]

}

df_sales_data = pd.DataFrame(sales_data)
print(df_sales_data)
```

```python
#ploat a line chart x,y

plt.figure(figsize=(12,4))
plt.plot(df_sales_data['Quarter'],df_sales_data['Item A'],color = 'red',label ='Item A')
plt.plot(df_sales_data['Quarter'],df_sales_data['Item B'],color ='blue',label = 'Item B')
plt.plot(df_sales_data['Quarter'],df_sales_data['Item C'],color = 'green' ,label ='Item C')
plt.title("Product sales")
plt.ylabel("Revenue (in lakh) ")

plt.xlabel("Financial Quarter")






plt.legend() #shoe the lebels
plt.show()

```
