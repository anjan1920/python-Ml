```python
#describe method
#its gives the smarmy of 

#let make a  data set and save 

herapheri_character = {
    'Name': ["Raju", "Shyam", "Babu bhaiya", "Tiwari", "Anuradha"],
    'Age': [30, 32, 45, 49, 29],
    'Catchphrase': [
        "25 din mein paisa double",               # Raju
        "Main time pe aata hoon!",              # Shyam
        "Utha le re baba!",                     # Babu bhaiya
        "Mera 40 lakh de!",                     # Tiwari
        "Paise ka bandobast ho gaya?"           # Anuradha
    ]
}

#make the data frame
df_herapheri_character = pd.DataFrame(herapheri_character)
print(df_herapheri_character)

#now use the describe method 
print(f"Descriptive statistics {df_herapheri_character.describe()}")

# Output -
# Descriptive statistics              Age
# count   5.000000
# mean   37.000000
# std     9.300538
# min    29.000000
# 25%    30.000000
# 50%    32.000000
# 75%    45.000000
# max    49.000000

# Though my data set have only one column of int values 
# the info gives the values related to this 
# like "age" the count -< noNull count -> 5 i.e all column are no null
# then the mean i.e the mean of age 
# and so on, std,max..
# .describe() in pandas gives summary stats for numeric columns.
# It internally **sorts the data i.e Age column** (not visibly) to calculate percentiles.
# 25%     -> 1st quartile (25% of values are ≤ this) ->based on sorted data
# 50%     -> Median (50% of values ≤ this) -> NOT based on row position
# 75%     -> 3rd quartile (75% of values ≤ this)

# First Quartile (25%) explained:
# When the Age column is sorted → [29, 30, 32, 45, 49]
# The 25% value (Q1) = 30
#  This means: 25% of the data values are **less than or equal to 30**
#  i.e., 30 is the value below which the first quarter of the sorted data lies.
# So, 29 and 30 are in the lowest 25% range (2 values out of 5 = 40%, but it's an approximate)
#
#  Think of quartiles as cutting sorted data into 4 equal parts.
#  - Q1 (25%) -> lower bound
#  - Q2 (50%) >median
#  - Q3 (75%) > upper bound of middle data
```
